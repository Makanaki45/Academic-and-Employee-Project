```python
pt.plot(employee.Department, employee['Number of Employees'])
pt.scatter(employee.Department, employee['Number of Employees'])
pt.xlabel("Deparment", color="blue")
pt.ylabel("Number of Employees", color="blue")
pt.title("Graph Showing Number of Employees per Department")

pt.show()
```

```python
pt.plot(employee.Department, employee['Total Salary/Department'])
pt.scatter(employee.Department, employee['Total Salary/Department'])
pt.xlabel("Department", color="Blue")
pt.ylabel("Total Salary Per Department", color="blue")

pt.title("Graph of Total Salary Per Department")
pt.show()
```

```python
pt.plot(employee.Department, employee['Total Commission/Department'])
pt.scatter(employee.Department, employee['Total Commission/Department'])
pt.xlabel("Department", color="Blue")
pt.ylabel("Total Commission Per Department", color="Blue")

pt.title("Graph of Total Commission Per Department")
pt.show()
```

```python
pt.plot(employee.Department, employee['Max Salary/Department'])
pt.scatter(employee.Department, employee['Max Salary/Department'])
pt.xlabel("Department", color="Blue")
pt.ylabel("Max Salary Per Department", color="Blue")

pt.title("Graph of Max Salary Per Department")
pt.show()
```

```python
pt.plot(employee.Department, employee['Min Salary/Department'])
pt.scatter(employee.Department, employee['Min Salary/Department'])
pt.xlabel("Department", color="Blue")
pt.ylabel("Min Salary Per Department", color="Blue")

pt.title("Graph of Min Salary Per Department")
pt.show()
```

```python
pt.bar(employee.Department, employee['Number of Employees'], color="Magenta")
pt.xlabel("Department", color="Magenta")
pt.ylabel("Number of Employees", color="Magenta")
pt.title("Graph Showing Number of Employees Per Department")

pt.show()
```

```python
pt.bar(employee.Department, employee['Total Salary/Department'])
pt.xlabel("Department", color="Blue")
pt.ylabel("Total Salary Per Department", color="Blue")

pt.title("Graph of Total Salary Per Department")
pt.show()
```

```python
pt.pie(employee['Number of Employees'], labels=employee.Department, autopct="%0.1f%%")
pt.title("Total Number of Employees Per Department")
pt.show()
```

```python
pt.pie(employee['Total Salary/Department'], labels=employee.Department, autopct="%0.2f%%")
pt.title("Total Salary Per Department")
pt.show()
```

```python
pt.subplot(1,2,1)
pt.bar(employee.Department, employee['Number of Employees'], color="Magenta")
pt.xlabel("Department", color="Magenta")
pt.ylabel("Number of Employees", color="Magenta")
pt.title("Graph Showing Number of Employees per Department")

pt.show()

pt.subplot(1,2,2)
pt.bar(employee.Department, employee['Total Salary/Department'])
pt.xlabel("Department", color="Blue")
pt.ylabel("Total Salary Per Department", color="Blue")

pt.title("Graph of Total Salary Per Department")
pt.show()
```
