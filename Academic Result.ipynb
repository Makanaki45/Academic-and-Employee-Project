```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Student_score.csv')
```

```python
df
```

```python
df1 = pd.read_csv('English Grade.csv')
```

```python
df1
```

```python
df = df.merge(df1)
```

```python
df
```

```python
df.drop(columns='Unnamed: 0', inplace=True)
```

```python
df
```

```python
df.info()
```

```python
df['Gender'] = df['Gender'].replace({'male': 'M', 'female':'F'})
```

```python
df
```

```python
df['City'] = df['City'].str.replace('St', 'Saint')
```

```python
df
```

```python
df['Total Score'] =df['Math Score'] + df['Science Score'] + df['English Score']
```

```python
df
```

```python
gender_group = df.groupby('Gender')
gender_mean = gender_group['Total Score'].mean()
gender_mean
```

```python
df['Average per Group'] = df.groupby('Gender')['Total Score'].transform('mean')
```

```python
df
```

```python
pivot_table = pd.pivot_table(df, values=['Math Score', 'English Score', 'Science Score'], index='Gender', aggfunc=np.mean, margins=True)
```

```python
pivot_table
```

```python
df = df.rename(columns={'Total Score':'Score'})
```

```python
df
```

```python
df['Score Above Mean'] = np.mean(df['Score'] >= df['Average per Group'], 'Yes', 'No')
```
